# 1. Generating data

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def dNdt(N,t):
    return R*N*(1-N/K)

R = 2
K = 10

N0 = 1
ts = np.linspace(0,10,1000)
Ns = odeint(dNdt,N0,ts)

print(Ns.shape)

ts_sample = ts[0:500:50]
np.random.seed(0)
Ns_sample = Ns[0:500:50,0] + 0.5*np.random.standard_normal(size=len(ts_sample))

plt.plot(ts,Ns)
plt.plot(ts_sample,Ns_sample,'o')
plt.xlabel('t')
plt.ylabel('N')
plt.show()
```

# 2. Simple feedforward neural net

```python
import torch
import torch.nn as nn
import torch.optim as optim

model = nn.Sequential(
    nn.Linear(1, 10),
    nn.ReLU(),
    nn.Linear(10, 10),
    nn.ReLU(),
    nn.Linear(10, 1),
    nn.ReLU(),
)

loss_fn = nn.MSELoss()  # mean squared error
optimizer = optim.Adam(model.parameters(), lr=0.001)

t_data = torch.tensor(ts_sample, dtype=torch.float32).reshape(-1, 1)
N_data = torch.tensor(Ns_sample, dtype=torch.float32).reshape(-1, 1)

N_EPOCHS = 5000
for epoch in range(N_EPOCHS):
    N_pred = model(t_data)
    loss = loss_fn(N_pred, N_data)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    #print('loss:',loss.item())
```

```python
ts_test = np.linspace(0,10,100)
ts_test_torch = torch.tensor(ts_test, dtype=torch.float32).reshape(-1, 1)
Ns_model = model(ts_test_torch).detach().cpu().numpy()
plt.plot(ts,Ns,'--')
plt.plot(ts_test,Ns_model)
plt.plot(ts_sample,Ns_sample,'o')
plt.legend(['True solution','Model preduction','Noisy training data'])
```
