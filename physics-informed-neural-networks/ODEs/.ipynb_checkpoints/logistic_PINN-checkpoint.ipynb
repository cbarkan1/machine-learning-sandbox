## Training a PINN with noisy data

ODE:  $dy/dt = Ry(1-y/K)$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def dNdt(N,t):
    return R*N*(1-N/K)

R = 2
K = 10

N0 = .1
ts = np.linspace(0,10,1000)
Ns = odeint(dNdt,N0,ts)
```

```python
# Plot noisy data and true solution

ts_sample = ts[0:500:50]
np.random.seed(1)
Ns_sample = Ns[0:500:50,0] + 0.5*np.random.standard_normal(size=len(ts_sample))
Ns_sample[Ns_sample<0] = 0 # All data should be non-negative

plt.plot(ts,Ns,'--')
plt.plot(ts_sample,Ns_sample,'o')
plt.xlabel('t')
plt.ylabel('N')
plt.legend(['True solution','Noisy training data'])
plt.show()
```

## 1. Simple feedforward neural net (not a PINN)

```python
import torch
import torch.nn as nn
import torch.optim as optim

model = nn.Sequential(
    nn.Linear(1, 20),
    nn.GELU(),
    nn.Linear(20, 20),
    nn.GELU(),
    nn.Linear(20, 1),
    nn.GELU(),
)

loss_fn = nn.MSELoss()  # mean squared error
optimizer = optim.Adam(model.parameters(), lr=0.001)

t_data = torch.tensor(ts_sample, dtype=torch.float32).reshape(-1, 1)
N_data = torch.tensor(Ns_sample, dtype=torch.float32).reshape(-1, 1)

N_EPOCHS = 15000
for epoch in range(N_EPOCHS):
    N_pred = model(t_data)
    loss = loss_fn(N_pred, N_data)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    #print('loss:',loss.item())
```

```python
ts_test = np.linspace(0,10,100)
ts_test_torch = torch.tensor(ts_test, dtype=torch.float32).reshape(-1, 1)
Ns_model = model(ts_test_torch).detach().cpu().numpy()
plt.plot(ts,Ns,'--')
plt.plot(ts_sample,Ns_sample,'o')
plt.plot(ts_test,Ns_model)
plt.legend(['True solution','Noisy training data','Model preduction'])
```

**The model severely overfits the data, and does not generalize
outside of the range of training data.**

## 2. PINN

```python
import torch.nn.functional as F

# Defining the physics loss function (mean squared deviation from ODE)
ts_test_with_grad = torch.tensor(ts_test, dtype=torch.float32, requires_grad=True).reshape(-1, 1)
def ODE_loss(model):
    Ns_model = model(ts_test_with_grad)
    dNdt_model = torch.autograd.grad(outputs=Ns_model, inputs=ts_test_with_grad, 
                                     grad_outputs=torch.ones_like(Ns_model), create_graph=True)[0]
    dNdt_true = dNdt(Ns_model,ts_test)
    return torch.mean((dNdt_true - dNdt_model)**2)

# Total loss
def loss_fn(model, inputs, target):
    outputs = model(inputs)
    mse_loss = F.mse_loss(outputs, target)
    ode_loss = ODE_loss(model)
    return mse_loss + ode_loss

N_EPOCHS = 10000
for epoch in range(N_EPOCHS):
    loss = loss_fn(model, t_data, N_data)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    #print('loss:',loss.item())
```

```python
ts_test = np.linspace(0,10,100)
ts_test_torch = torch.tensor(ts_test, dtype=torch.float32).reshape(-1, 1)
Ns_model = model(ts_test_torch).detach().cpu().numpy()
plt.plot(ts,Ns,'--')
plt.plot(ts_sample,Ns_sample,'o')
plt.plot(ts_test,Ns_model)
plt.legend(['True solution','Noisy training data','Model preduction'])
```

**Now, the model closely matches the true solution.**
